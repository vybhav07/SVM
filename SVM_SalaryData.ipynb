```python
from sklearn.preprocessing import LabelEncoder

encoding_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']

encodLable=LabelEncoder()
for i in encoding_col:
    salary_train[i]=encodLable.fit_transform(salary_train[i])
    salary_test[i]=encodLable.fit_transform(salary_test[i])
```

```python
salary_train.head()
```

```python
salary_test.head()
```

```python
def norm(i):
    return (i-i.min())/(i.max()-i.min())
```

```python
x_train=salary_train.iloc[:,0:13]
y_train=salary_train.iloc[:,13]
x_test=salary_test.iloc[:,0:13]
y_test=salary_test.iloc[:,13]
```

```python
x_train=norm(x_train)
x_test=norm(x_test)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# Grid Search

clf = SVC()
param_grid = [{'kernel':['linear','rbf','polynomial'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.01] }]
gsv = RandomizedSearchCV(clf,param_grid,cv=6,) 
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(kernel='linear',C=10)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
